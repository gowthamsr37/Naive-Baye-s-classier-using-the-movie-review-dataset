# Text Cleaning
- 1.Sample 10000 rows
- 2.Remove html tags
- 3.Remove special characters
- 4.Convert to lower case
- 5.Removing stop words
- 6.Stemming

```python
df  = df.sample(10000)
```

```python
df.shape
```

```python
df.info()
```

```python
df['sentiment'] = df['sentiment'].replace({'positive':1 , 'negative':0})
```

```python

df.reset_index(inplace=True , drop=True)
df.head()
```

```python
import re
```

```python
def clean_html(text):
    clean = re.compile('<.*?>')
    return re.sub(clean , '', text)
```

```python
df['review'] = df['review'].apply(clean_html)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
len(stopwords.words('english'))
```

```python
df['review'][0]
```

```python
corpus = []

for i in range(len(df['review'])):
    
    text = re.sub('[^a-zA-Z]' , ' ' , df['review'][i])
    text = text.lower()
    text = text.split()
    
    text = [ps.stem(word) for word in text if word not in set(stopwords.words('english'))]
    
    text = ' '.join(text)
    corpus.append(text)
```

```python
corpus[3]
```

```python
df['review'] = corpus
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
X= cv.fit_transform(df['review']).toarray()
```

```python
X
```

```python
X.shape
```

```python
cv.vocabulary_
```

```python
y = df.iloc[:,-1].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test ,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
print(X_train.shape) , print(X_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB , MultinomialNB , BernoulliNB
gnb = GaussianNB()
mnb = MultinomialNB()
bnb=BernoulliNB()
```

```python
gnb.fit(X_train, y_train)
mnb.fit(X_train, y_train)
bnb.fit(X_train, y_train)
```

```python
y_pred_gnb = gnb.predict(X_test)
y_pred_mnb = mnb.predict(X_test)
y_pred_bnb = bnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Gausian NB Accuracy: ', accuracy_score(y_test , y_pred_gnb))
print('Multinomial NB Accuracy: ', accuracy_score(y_test , y_pred_mnb))
print('Bernaulli NB Accuracy: ', accuracy_score(y_test , y_pred_bnb))
```
